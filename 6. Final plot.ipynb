```python
import matplotlib.pyplot as plt
from matplotlib import ticker
import numpy as np
```

```python
xlabels = ['Random Forest', 'AdaBoost', 'XGBoost'] # masukin list kernel nya yang di pake apa aja
ylabels = np.arange(start=0.9, stop=1, step=0.01)
```

```python
RF_Untuned = 93.28
AB_Untuned = 92.10
XB_Untuned = 94.58
RF_Tuned = 97.93
AB_Tuned = 96.37
XB_Tuned = 99.00
akurasi_1 = [RF_Untuned, AB_Untuned, XB_Untuned] # akurasi  di urutin sama list model
akurasi_2 = [RF_Tuned, AB_Tuned, XB_Tuned] # akurasi di urutin sama list model
diff = [(RF_Tuned-RF_Untuned),(AB_Tuned-AB_Untuned),(XB_Tuned-XB_Untuned)]
```

```python
x = np.arange(len(xlabels))  # yang ini ga usah di ubah
width = 0.2  # lebar bar
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
plt.gcf().set_dpi(1000)
bar1 = ax.bar(x - 0.21, akurasi_1, width, label='Untuned Accuracy', color='b')
bar2 = ax.bar(x + 0.0, akurasi_2, width, label='Tuned Accuracy', color='g')
bar3 = ax.bar(x + 0.21, diff, width, label='Difference', color='r')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy', fontsize = 15)
ax.set_xticks(x)
ax.set_xticklabels(xlabels, fontsize=12)
plt.yticks(fontsize=12)
ax.legend(fontsize=12, loc=6)

ax.bar_label(bar1, padding=5, fontsize=11)
ax.bar_label(bar2, padding=5, fontsize=11)
ax.bar_label(bar3, padding=5, fontsize=11)

plt.show()
```
