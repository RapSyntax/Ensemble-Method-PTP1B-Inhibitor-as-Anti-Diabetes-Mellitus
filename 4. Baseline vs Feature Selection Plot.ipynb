```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
xlabels = ['Random Forest', 'Adaptive Boosting', 'XGBoost']
ylabels = np.arange(start=0, stop=1, step=0.01)

accRF1 = 93.50
accAB1 = 91.10
accXG1 = 94.39
accRF2 = 93.28
accAB2 = 92.10
accXG2 = 94.58

akurasi_1 = [accRF1, accAB1, accXG1] #akurasi baseline
akurasi_2 = [accRF2, accAB2, accXG2] #akurasi setelah feature selection
selisih = [(accRF2-accRF1), (accAB2-accAB1), (accXG2-accXG1)]


x = np.arange(len(xlabels)) 
width = 0.2 

fig, ax = plt.subplots(figsize=(10,8))
plt.gcf().set_dpi(1000)
rects2 = ax.bar(x - 0.21, akurasi_1, width, label='Using Baseline Parameters and All Features', color='b')
rects3 = ax.bar(x + 0.0, akurasi_2, width, label='After Feature Selection', color='g')
rects4 = ax.bar(x + 0.21, selisih, width, label='Difference', color='r')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Accuracy', fontsize = 15)
ax.set_xticks(x)
ax.set_xticklabels(xlabels, fontsize=12)
plt.yticks(fontsize=12)
ax.legend(fontsize=12, loc=6)

ax.yaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(1))

ax.bar_label(rects2, padding=5, fontsize=11)
ax.bar_label(rects3, padding=5, fontsize=11)
ax.bar_label(rects4, padding=5, fontsize=11)

plt.show()
```
